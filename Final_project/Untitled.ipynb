```python
import pandas as pd

# Read in training data in chunks
chunk_size = 100
train_dataframe_iterator = pd.read_csv('train1.csv', chunksize=chunk_size)

# Preprocess data
def preprocess_data(dataframe):
    input_texts = []
    summary_texts = []
    for row in tqdm(dataframe.iterrows()):
        input_text = row[1]['q']
        summary_text = row[1]['target']
        # Tokenize input and summary text
        tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
        input_tokens = tokenizer.encode(input_text, return_tensors='pt')
        summary_tokens = tokenizer.encode(summary_text, return_tensors='pt')
        input_texts.append(input_tokens)
        summary_texts.append(summary_tokens)
    return input_texts, summary_texts

# Process data in chunks
input_texts = []
summary_texts = []
for dataframe in train_dataframe_iterator:
    input_texts_chunk, summary_texts_chunk = preprocess_data(dataframe)
    input_texts.extend(input_texts_chunk)
    summary_texts.extend(summary_texts_chunk)
```

```python
def generate_summary(input_text, model, tokenizer, max_length=100):
    # Encode input text and add special tokens
    input_ids = tokenizer.encode(input_text, return_tensors='pt')
    input_ids = torch.cat((input_ids, torch.tensor([[2]])), dim=1)
    # Generate summary
    summary_ids = model.generate(input_ids, max_length=max_length)
    # Remove special tokens
    summary_ids = summary_ids[:, 1:]
    # Decode summary
    summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary_text
```
